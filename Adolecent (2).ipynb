# ## Separating independent(x) & Dependent(y) value

```python
from sklearn.model_selection import train_test_split
x = df2.drop(['CLASS/ASD TRAITS_YES'], axis=1)
y = df2['CLASS/ASD TRAITS_YES']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x
```

```python
y
```

# Now Training Model

Logistic Reegression

```python
#logistic method
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=1000, C=1.0)
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
logreg.score(x_train, y_train)
```

```python
y_pred
```

```python
# calculate R2 score
from sklearn.metrics import r2_score
logreg_r2 = r2_score(y_test, y_pred)
print('R2 score:',logreg_r2)
```

```python
#classification report
#confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
c_report = classification_report(y_test, y_pred)
print('Classification Report\n', c_report )
```

```python
confusion_matrix(y_test,y_pred)
```

##DECISION TREE

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
y_pred = dtree.predict(x_test)
dtree.score(x_train, y_train)
```

```python
# calculate R2 score
dtree_r2 = r2_score(y_test, y_pred)
print('R2 score:',dtree_r2)
```

```python
#classification report
#confusion matrix
c_report = classification_report(y_test, y_pred)
print('Classification Report\n', c_report )
```

```python
confusion_matrix(y_test,y_pred)
```
